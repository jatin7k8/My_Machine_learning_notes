# Euclidean distance(l2 Norm) :-
* works on the principle of the Pythagoras theorem between two points.
* calculates the distance between two real-valued vectors.

image.png

# By python

```python
from math import sqrt

# calculate euclidean distance
def euclidean_distance(a, b):
 return sqrt(sum((e1-e2)**2 for e1, e2 in zip(a,b)))

# define data
row1 = [10, 20, 15, 10, 5]
row2 = [12, 24, 18, 8, 7]
# calculate distance
dist = euclidean_distance(row1, row2)
print(dist)
```

# Scipy implementation

```python
# calculating euclidean distance between vectors
from scipy.spatial.distance import euclidean
# define data
row1 = [10, 20, 15, 10, 5]
row2 = [12, 24, 18, 8, 7]
# calculate distance
dist = euclidean(row1, row2)
print(dist)
```

## uses of Eucliden Distance :-
**1. Clustering:**

* **Algorithm:** K-means clustering

* **Use of Euclidean Distance:**
In K-means clustering, built-in points known as centroids represent cluster centers. Euclidean distance is calculated between each data point and all centroids. The data point is then assigned to the cluster represented by the closest centroid.

* **Real-life Example:** Retail store layout optimization. Imagine a retail chain wants to optimize its store layouts based on customer behavior. Customer purchase data can be clustered using K-means with built-in centroids representing different store sections (e.g., clothing, electronics). Euclidean distance helps determine which section each customer's purchase behavior is closest to, enabling the store to tailor layouts for maximum sales.









**2. Classification:**

* **Algorithm:** K-nearest neighbors (KNN)

* **Use of Euclidean Distance:** In KNN classification, labeled data points serve as built-in points. Euclidean distance is computed between a new data point and all labeled points in the training set. The class labels of the k-nearest neighbors are used to predict the class label of the new data point.

* **Real-life Example:** Online movie recommendation. Suppose a streaming platform recommends movies based on user preferences. KNN with Euclidean distance can be used to find similar users (built-in points) based on their movie-watching history. The platform then recommends movies liked by similar users to the target user.

**3. Dimensionality Reduction:**

* **Algorithm:** Principal Component Analysis (PCA)

* **Use of Euclidean Distance:** PCA uses built-in points as reference vectors in the original feature space. Euclidean distance measures the similarity between data points and these reference vectors, facilitating dimensionality reduction while preserving variance.

* **Real-life Example:**Genetics and gene expression analysis. In genetics, PCA can be applied to gene expression data to identify patterns across different samples. Built-in points represent reference gene expression profiles, and Euclidean distance helps identify similar gene expression patterns, aiding in understanding genetic variations and disease mechanisms.

**4. Recommendation Systems:**

* **Algorithm:** Collaborative filtering

* **Use of Euclidean Distance:** Collaborative filtering uses built-in points to represent users or items with known preferences. Euclidean distance compares the preferences of a target user or item with those of all built-in points, enabling personalized recommendations.

* **Real-life Example:** E-commerce product recommendations. Consider an online marketplace recommending products to users. Collaborative filtering with Euclidean distance can be employed to find similar users based on their purchase history. Products preferred by similar users are then recommended to the target user.

**5. Anomaly Detection:**
* **Algorithm:** Anomaly detection using distance-based methods

* **Use of Euclidean Distance:** Built-in points serve as reference points for normal behavior. Euclidean distance is calculated between data points and all built-in points. Data points significantly distant from any of the built-in points may indicate anomalies or outliers.

* **Real-life Example:** Network intrusion detection. In cybersecurity, anomaly detection is used to identify unusual network traffic patterns indicating potential attacks. Euclidean distance measures the deviation of network traffic data from normal behavior represented by built-in points, helping detect suspicious activities.

## Limitations :-
1. Sensitivity to Scale.
2. Curse of Dimensionality.
3. Assumption of Linear Relationships.
4. Sensitivity to Outliers.
5. Not Robust to Irrelevant Features.
6. Limited Applicability to Non-Euclidean Spaces.
7. Computational Complexity.
8. Assumption of Continuity.

**1. Sensitivity to Scale:-**  
* Euclidean distance is influenced by the scale of the variables.
* Features with larger scales can dominate the distance calculation.
* Normalization or standardization techniques are often necessary to ensure fair comparisons.

Example: Consider a dataset with two features: height (measured in meters) and weight (measured in kilograms). The scale of weight is much larger than height. Without normalization or standardization, Euclidean distance calculations would be dominated by weight, potentially leading to biased results

**2. Curse of Dimensionality:**
* In Lower high-dimensional spaces, it will work effectively
* In high-dimensional spaces, Euclidean distance becomes less meaningful.
* All points tend to be roughly equidistant from each other.
* This can significantly impact the performance of algorithms reliant on distance metrics, such as KNN or clustering.

Example: In a dataset with 100 features (dimensions), all uniformly distributed between 0 and 1, the average distance between data points becomes large as the number of dimensions increases. Thus, in high-dimensional spaces, Euclidean distance may not accurately reflect the true relationships between data points.


**3. Assumption of Linear Relationships:**
* Euclidean distance assumes linear relationships between variables.
* However, many real-world datasets contain nonlinear relationships.
* This assumption can lead to inaccuracies in similarity assessments between data points.

Example:-  Suppose we have a dataset containing variables such as age and income. While age may have a linear relationship with income up to a certain point, it may not hold true for all age groups. Euclidean distance may inaccurately measure the dissimilarity between individuals with similar incomes but different ages.

**4.Sensitivity to Outliers:**
* Euclidean distance is sensitive to outliers.
* Outliers can distort the distance calculation and skew results.
* They may lead to inaccuracies in clustering or classification tasks.

Example: In a dataset of housing prices, an outlier representing an extremely expensive mansion may significantly affect the distance calculations between data points. This outlier could distort the clustering results, leading to inaccurate groupings of similar-priced houses.

**5. Not Robust to Irrelevant Features:**
* Euclidean distance treats all features equally, irrespective of their relevance.
* Irrelevant or noisy features can negatively impact distance calculations.
* This can result in suboptimal model performance or misleading insights.

Example: Consider a dataset containing customer information for marketing analysis. If the dataset includes irrelevant features such as customer IDs or timestamps, Euclidean distance calculations would be influenced by these irrelevant features, potentially leading to suboptimal clustering or classification results.

**6. Limited Applicability to Non-Euclidean Spaces:**
* Euclidean distance is well-defined only in Euclidean spaces.
* It may not be suitable for domains requiring alternative distance metrics, such as spherical data or graph data.
* Non-Euclidean data may require specialized distance measures for accurate analysis.

Example: Suppose we have geographic data represented as latitude and longitude coordinates. Calculating Euclidean distance between two points on the Earth's surface would not accurately reflect the actual distances between those points, as the Earth's surface is spherical. Instead, a distance metric like the Haversine formula would be more appropriate.

**7. Computational Complexity:**
* Calculating Euclidean distance for all pairs of data points can be computationally expensive.
* Especially problematic for large datasets with numerous dimensions.
* This can lead to scalability issues, particularly in real-time or resource-constrained environments.

Example: In a dataset with a million data points and a thousand features each, calculating pairwise Euclidean distances would involve computing a billion distance calculations. This computation can be prohibitively expensive, especially in real-time applications or with limited computational resources.

**8.Assumption of Continuity:**
* Euclidean distance assumes that the underlying space is continuous.
* It may not be appropriate for discrete or categorical data.
* In such cases, alternative distance metrics tailored to the data type may be more suitable.

Example: Consider a dataset containing categorical variables such as color or product categories. Euclidean distance, which assumes a continuous space, would not be meaningful for comparing these categorical variables. Instead, techniques such as one-hot encoding or alternative distance metrics tailored to categorical data should be used.

# Q. What is the difference between Euclidean and Mahalanobis metrics for distance calculation, especially when applied to clustering problems?


# solution

The Euclidean distance and Mahalanobis distance are both measures used to calculate the distance between two points, but they do so in significantly different ways, which affects their effectiveness in various scenarios, such as clustering.

Euclidean Distance

Euclidean distance is the most common use of distance in geometry, often referred to as the “ordinary” distance between two points in Euclidean space. It is calculated by taking the square root of the sum of the squares of the differences between corresponding coordinates of the two points. Mathematically, the Euclidean distance between two points A and B with coordinates (a1, a2, …, an) and (b1, b2, …, bn) in an n-dimensional space is given by:

d_Euclidean(A, B) = sqrt((a1 - b1)^2 + (a2 - b2)^2 + … + (an - bn)^2)

Mahalanobis Distance

The Mahalanobis distance, on the other hand, is a measure of the distance between a point and a distribution. Unlike the Euclidean distance, which simply considers the geometric distance, the Mahalanobis distance takes into account the correlations of the data set and is scale-invariant, meaning the scale of the measurements does not affect the distance calculation. The Mahalanobis distance between a point x and a distribution with mean μ and covariance matrix S is given by:

d_Mahalanobis(x, μ) = sqrt((x - μ)T S^(-1) (x - μ))

Comparison and Context in Clustering

 • Scale and Correlation Sensitivity: Euclidean distance treats all dimensions equally and does not account for correlations between dimensions. This can be problematic in clustering when different features have different scales or units, or when the data is not isotropically distributed. Mahalanobis distance, by considering the covariance matrix, naturally adjusts for scale and correlation, making it more accurate for measuring “true” distances in such cases.
 • Handling Outliers: For points that are outside the typical data distribution, the Mahalanobis distance’s accounting for the data distribution makes it more robust. It can more accurately reflect the atypicality of these points in the context of the overall dataset, as opposed to Euclidean distance which might not accurately reflect their distance.
 • Clustering: In clustering, using Mahalanobis distance can lead to clusters that more accurately reflect the underlying data distribution. This is particularly true in cases where the clusters are elliptical rather than spherical, or when they have varying sizes and densities.

In summary, while Euclidean distance is straightforward and effective for uniform, scale-independent data, Mahalanobis distance offers advantages when dealing with heterogeneously scaled data & data with correlated features.
