# Manhattan Distance (Taxicab or City Block Distance)
* Also called as rectilinear distance, city block distance, taxicab metric, L1 Norm
* Manhattan distance is calculated as the sum of the absolute differences between the two vectors.
* Less sensitive to outliers than Euclidean distance.
* Computationally simpler, especially in high-dimensional spaces.
* Can be a good choice for sparse or categorical data.

image.png

* calculates the total distance traveled along a grid-like path, considering only horizontal and vertical movements. It's the sum of the absolute differences between corresponding coordinates of two data points.

# Scipy implemantation


```python
from scipy.spatial.distance import cityblock
# define data
row1 = [10, 20, 15, 10, 5]
row2 = [12, 24, 18, 8, 7]
# calculate distance
dist = cityblock(row1, row2)
print(dist)
```

# Uses of Manhattan Distance:

**1. Clustering:**
* Manhattan distance can be used in clustering algorithms like K-means to group data points based on their similarity.

**2. Classification:**
* Similar to Euclidean distance, Manhattan distance can be applied in classification algorithms such as K-nearest neighbors (KNN) to predict the class label of a new data point based on its proximity to labeled points.

**3. Dimensionality Reduction:**
* In techniques like t-distributed stochastic neighbor embedding (t-SNE),
*  Manhattan distance can help measure the similarity between high-dimensional data points when reducing their dimensionality for visualization.

**4. Recommendation Systems:**
* Manhattan distance can be used in collaborative filtering-based recommendation systems to find similar users or items based on their preferences.

**5. Route Planning:**
* In navigation applications, Manhattan distance is often used to calculate the distance between two points on a grid-like city map where movement is restricted to horizontal and vertical paths.

# Python Implementation


```python
def manhattan_distance(a, b):
	return sum(abs(e1-e2) for e1, e2 in zip(a,b))

# define data
row1 = [10, 20, 15, 10, 5]
row2 = [12, 24, 18, 8, 7]
# calculate distance
dist = manhattan_distance(row1, row2)
print(dist)
```

#In higher dimensions, Manhattan Distance is often preferred over Euclidean Distance for a few key reasons:



**1. Curse of Dimensionality:**
* As the dimensionality increases, the Euclidean distance between points tends to increase proportionally.
* This means that in high-dimensional spaces, distances between points become less meaningful and can lead to poor performance of algorithms that rely on distance calculations, such as clustering or nearest neighbor search.

**2. Sparse Data:**
* In high-dimensional spaces, data often becomes sparse, meaning that points are far apart from each other in most dimensions.
* In such cases, Manhattan Distance, which measures the distance along each dimension independently, can be more suitable than Euclidean Distance, which considers the straight-line distance between points.

**3. Computational Efficiency:**
*  Manhattan Distance is computationally cheaper to calculate compared to Euclidean Distance, especially in high-dimensional spaces.
* This is because Manhattan Distance only involves summing the absolute differences along each dimension, while Euclidean Distance requires taking square roots and summing squared differences.

**4. Robustness to Outliers:**
* Manhattan Distance is less sensitive to outliers compared to Euclidean Distance.
* In high-dimensional spaces, outliers can have a significant impact on distance calculations, and Manhattan Distance's focus on the differences along each dimension can mitigate this effect.

# Limitations of Manhattan Distance

* Not Always Optimal: While Manhattan distance can be effective in specific scenarios, it may not always be the best choice for measuring distance. Unlike Euclidean distance, it doesn't account for the straight-line path, potentially leading to inaccurate results for certain data distributions.

* Sensitive to Outliers: Similar to Euclidean distance, Manhattan distance can be sensitive to outliers, as large differences in a single dimension can significantly impact the overall distance.

* Ignores Correlations: Manhattan distance treats all dimensions independently, not considering potential correlations between features. This can be a disadvantage compared to metrics that incorporate correlations.
